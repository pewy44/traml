```python
import numpy as np
```

```python
def fun(x, stack, film_range, c, delta):
    for i in range(film_range[0],film_range[-1]):
        stack.d(i,x[i])
    #stack.d(1,x[0])
    #stack.d(2,x[1])
    for p in x:
        p = np.sqrt(x[0]**2)
    #x[0]=np.sqrt(x[0]**2)
    min = 1-stack.average('R', c, delta)
    #print min, x[0], x[1]

    return min**10
```

```python
def minTR(stack, film_range, c, delta):
    x = []
    layers = []
    bnds = []
    for i in range(film_range[0],film_range[-1]+1):
        x.append(stack.config[i][1])
        layers.append(i)
        bnds.append((5,500))
        
    minimize(fun, x, args=(a, layers, c,delta), bounds = bnds,tol=1e-30, method='SLSQP')
    stack.plot()
    return bnds
```

```python
minTR(a,[1,10],0.4,0.01)
```
